```python
import pandas as pd
import sqlite3
```

```python
df = pd.read_csv("netflix_titles.csv")
```

```python
df
```

```python
df.info()
```

```python
for i in df.columns:
    null_rate = df[i].isna().sum() / len(df) * 100 
    if null_rate > 0 :
        print("{} null rate: {}%".format(i,round(null_rate,2)))
```

```python
df.drop(["director", "cast", "date_added", "description"], axis=1, inplace=True)
```

```python
df
```

```python
df.info()
```

```python
df["country"] = df["country"].fillna("unknown")
df["rating"] = df["rating"].fillna("unknown")
df["duration"] = df["duration"].fillna("unknown")
```

```python
df.info()
```

```python
df.to_csv('netflix_titles_edited.csv', index=False)
```

```python
df
```

```python
conn = sqlite3.connect('netflix_titles.db')
```

```python
create_sql = "CREATE TABLE IF NOT EXISTS shows_and_movies (show_id INTEGER PRIMARY KEY, type TEXT, title TEXT, country TEXT, release_year TEXT, rating TEXT, duration TEXT, listed_in TEXT)"
```

```python
cursor = conn.cursor()
```

```python
cursor.execute(create_sql)
```
